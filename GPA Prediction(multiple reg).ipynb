```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
raw_data = pd.read_csv("./machine learning datasets/1.03. Dummies.csv")
```

```python
raw_data
```

```python

# usually we will map the yes and no to 1 and 0
# if the students attended more that 75%, YES =1 else No = 0
data = raw_data.copy()
```

```python
data["Attendance"] = data["Attendance"].map({"Yes":1, "No":0})
```

```python
data
```

```python
data.describe()
```

```python
#the mean is <0.5 which shows there are more 0s than 1s
# that is 46% students attended 


```

```python
# Regression
y = data["GPA"]
x1 = data[["SAT", "Attendance"]]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
```

summary
GPA = 0.6439 + 0,0014* SAT + 0.2226*Dummy
if students attend: GPA = 0.6439 + 0,0014* SAT + 0.2226*1
is students did not: GPA = 0.6439 + 0,0014* SAT + 0.2226*0

```python
plt.scatter(data["SAT"], y)
yhat_No = 0.6439 + 0.0016*data["SAT"]
yhat_yes = 0.8665 + 0.0016*data["SAT"]
fig = plt.plot(data["SAT"], yhat_No,lw=2, c = "#006837")
fig = plt.plot(data["SAT"], yhat_yes,lw=2, c = "#a50026")
plt.xlabel("SAT", fontsize = 20) 
plt.ylabel("GAP", fontsize = 20)
plt.show() 


```

attendance has red line and absence has blue line

```python
# Add regression line to the plot
plt.scatter(data["SAT"], y, c=data["Attendance"], cmap="RdYlGn_r")
yhat_No = 0.6439 + 0.0016*data["SAT"]
yhat_yes = 0.8665 + 0.0016*data["SAT"]
fig = plt.plot(data["SAT"], yhat_No,lw=2, c = "#006837", label="regression line1")
fig = plt.plot(data["SAT"], yhat_yes,lw=2, c = "#a50026", label="regression line2")
plt.xlabel("SAT", fontsize = 20) 
plt.ylabel("GAP", fontsize = 20)
plt.show() 
```
